```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
iris = sns.load_dataset("iris")
```

1: Violin Plot - Distribution of Sepal Length across Species

```python
plt.figure(figsize=(10, 6))
sns.violinplot(x="species", y="sepal_length", data=iris)
plt.title("Violin Plot: Sepal Length by Species")
plt.show()
```

 The violin plot shows that Setosa has shorter sepal lengths compared to Versicolor and Virginica. Virginica generally has the longest sepals, while Versicolor overlaps with both other species.

 2: Swarm Plot - Sepal Width across Species

```python
plt.figure(figsize=(10, 6))
sns.swarmplot(x="species", y="sepal_width", data=iris)
plt.title("Swarm Plot: Sepal Width by Species")
plt.show()
```

 Setosa has wider sepals than the other species, whereas Virginica tends to have the narrowest. There's a significant overlap between Versicolor and Virginica in terms of sepal width.

3: Strip Plot - Petal Length across Species

```python
plt.figure(figsize=(10, 6))
sns.stripplot(x="species", y="petal_length", data=iris)
plt.title("Strip Plot: Petal Length by Species")
plt.show()
```

The strip plot shows that Setosa consistently has the shortest petal lengths, while Virginica has the longest. Versicolor shows intermediate petal lengths, overlapping with both Setosa and Virginica.

4: Box Plot - Petal Width across Species

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x="species", y="petal_width", data=iris, width=.3, color = "cyan")
plt.title("Box Plot: Petal Width by Species")
plt.show()
```

Setosa has much narrower petals than the other species, while Virginica has the broadest. Versicolor overlaps with both but leans closer to Virginica.

5: Bar Plot - Mean Sepal Length by Species

```python
plt.figure(figsize=(10, 6))
sns.barplot(x="species", y="sepal_length", data=iris,width = 0.35, color = 'violet')
plt.title("Bar Plot: Mean Sepal Length by Species")
plt.show()
```

On average, Virginica has the longest sepals, followed by Versicolor and then Setosa.

6: Count Plot - Distribution of Species

```python
plt.figure(figsize=(8, 6))
sns.countplot(x="species", data=iris, width = .45)
plt.title("Count Plot: Distribution of Species")
plt.show()
```

The count plot shows that the dataset is well-balanced, with 50 samples from each of the three species.

7: Heatmap - Correlation Between Features

```python
plt.figure(figsize=(10, 6))
sns.heatmap(iris.drop(columns="species").corr(), annot=True, cmap="PuOr")
plt.title("Heatmap: Correlation Between Features")
plt.show()
```

he heatmap highlights a strong positive correlation between petal length and petal width (0.96), and between sepal length and petal length (0.87). Sepal width shows weaker correlations with the other features.

 8: Scatter Plot - Sepal Length vs Petal Length

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x="sepal_length", y="petal_length", hue="species", data=iris)
plt.title("Scatter Plot: Sepal Length vs Petal Length")
plt.show()
```

The scatter plot reveals distinct clusters for each species. Setosa is clearly separated from the others, while Versicolor and Virginica show some overlap, particularly in the middle ranges.


9: Line Plot - Sepal Length across Sorted Samples

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x=range(len(iris)), y=sorted(iris["sepal_length"]), hue=sorted(iris["species"]))
plt.title("Line Plot: Sepal Length across Sorted Samples")
plt.show()
```

The line plot shows a clear increase in sepal length from Setosa to Virginica, with Setosa having the shortest sepals and Virginica the longest.

10: Pair Plot - Pairwise Relationships Between Features

```python
sns.pairplot(iris, hue="species", diag_kind="kde")
plt.suptitle("Pair Plot: Pairwise Relationships Between Features", y=1.02)
plt.show()
```

The pair plot visually confirms that Setosa is distinct in most comparisons, while Versicolor and Virginica overlap more, indicating that Setosa is the most distinct of the three species.
